```python
import pandas as pd
import numpy as np
from numpy import arange
import matplotlib
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
```

## load and plot data

```python
df=pd.read_csv('biochar_data.csv',index_col=0)
df.sort_values('Temp (C)',inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
matplotlib.rcParams.update({'font.weight':'normal','font.size': 14})
```

```python
fig, axs = plt.subplots(nrows=20,ncols=1,figsize=[8,100])

for x, col in enumerate(list(df.columns[1:])):
    axs[x].scatter(df['Temp (C)'],df[col],marker='x',c=df['Temp (C)'],cmap='plasma')
    axs[x].set_xlabel('HTT (ºC)')
    axs[x].set_ylabel(col)

plt.show()
fig.savefig('plotted_data.pdf')
```

## logistic data (sigmoid)

```python
#define logistic function (sigmoid)
def logistic(x, L=1, k=1, x_0=0, b=0):
    y = L / (1 + np.exp(-k*(x - x_0)))+ b
    return y

#define sigmoid function w fitted curve
def fittedlogistic(x, fL=1, fb=0):
    y = fL / (1 + np.exp(-fk*(x - fx_0)))+ fb
    return y   
```

```python
#properties with logistic shape and initial guesses for p_0 params
logistic_properties=['True density (gcm-3)', 'H/C', 'Yield (%)', 'C (%)', 'H (%)', 'O (%)', 'O/C', 'NMR aromaticity (%)', 'VM (%)', 'FC (%)']
initial_guesses=[0.0155, -0.023, -0.0254, 0.0035, -0.001, -0.008, -0.0141, 0.0224, -0.0034, -0.0188] #these were done iteratively from popt
initial_guesses_dic={logistic_properties[i]:initial_guesses[i] for i in range(len(logistic_properties))}

#make dictionaries for parameters
popt_dic={prop:[] for prop in logistic_properties}
#make dictionaries for confidence limits
popt_upr_dic={prop:[] for prop in logistic_properties}
popt_lwr_dic={prop:[] for prop in logistic_properties}
```

```python
print(logistic_properties)
len(logistic_properties)
```

### fitting with means, lwr_bound and upr_bound

```python
fig, axs = plt.subplots(nrows=10,figsize=[8,55])

for i, prop in enumerate(logistic_properties):
    print(prop)
    #make temp df
    df_temp=df[["Temp (C)", prop]].dropna()
    #make temp means, standard devs and standard errors df
    df_stats = df_temp.groupby("Temp (C)").agg(mean=(prop,'mean'), std=(prop,'std'), count=(prop,'count'), lwr_bound=(prop, lambda val:np.quantile(val, q=0.01)), upr_bound=(prop, lambda val:np.quantile(val, q=0.99))).dropna()
    #x, y and sigma data for input to curve fit
    x = df_stats.index
    y = df_stats['mean']
    sig = df_stats['std']
    #initial guesses
    p0 = [max(y) - min(y), initial_guesses_dic[prop], np.median(x), min(y)]
    #fit data
    popt_dic[prop], _pcov = curve_fit(logistic, x, y, p0=p0, method='lm')
    _fL, fk, fx_0, _fb = popt_dic[prop]
    
    #find lower bounds
    lwr_y = df_stats['lwr_bound']
    popt_lwr_dic[prop], _pcov = curve_fit(fittedlogistic, x, lwr_y, p0=[_fL, _fb], method='lm')
    #find upper bounds
    upr_y = df_stats['upr_bound']
    popt_upr_dic[prop], _pcov = curve_fit(fittedlogistic, x, upr_y, p0=[_fL, _fb], method='lm')
    
    #x-range for lobf and confidence intervals
    x_min=df_temp['Temp (C)'].min()
    x_max=df_temp['Temp (C)'].max()
    x_range=np.arange(x_min,x_max,30)
    
    #plot
    axs[i].plot(df_temp['Temp (C)'], df_temp[prop], marker=".", linewidth=0, label='all data')
    axs[i].plot(x_range, logistic(x_range, *popt_dic[prop]), linestyle="--", color="r", label='fit of all data')
    axs[i].fill_between(x_range, fittedlogistic(x_range, *popt_lwr_dic[prop]), fittedlogistic(x_range, *popt_upr_dic[prop]), alpha=0.25, label='confidence interval')
    axs[i].legend()
    axs[i].set_ylabel(prop)
    axs[i].set_xlabel('HTT (ºC)')

plt.show()
fig.savefig('logistic_fittedplots.png')
```

## quadratic

```python
#define quadratic function
def quadratic(x, a=1):
    y = a* x**2
    return y

quadratic_bounds=(0,np.inf)

# potentially might want to include a loc and bias value to this quadratic function too!
def quadratic2(x, a=1, b=0):
    y = a*(x)**2 + b
    return y

quadratic2_bounds=((0,0),(np.inf,np.inf))
```

```python
#properties with quadratic shape
quadratic_properties=['SA (m2g-1)','Ash (%)','NMR delta (ppm)']

#make dictionaries for parameters and errors
popt_dic2={prop:[] for prop in quadratic_properties}
#make dictionaries for confidence limits
popt_upr_dic2={prop:[] for prop in quadratic_properties}
popt_lwr_dic2={prop:[] for prop in quadratic_properties}
```

```python
print(quadratic_properties)
```

### using means, lwr_bound and upr_bound

```python
fig, axs = plt.subplots(nrows=len(quadratic_properties),figsize=[8.7,len(quadratic_properties)*5.5])

for i, prop in enumerate(quadratic_properties):  
    df_temp=df[['Temp (C)',prop]].dropna()
    #make temp means, standard devs and standard errors df
    df_stats = df_temp.groupby("Temp (C)").agg(mean=(prop,'mean'), std=(prop,'std'), count=(prop,'count'), lwr_bound=(prop, lambda val:np.quantile(val, q=0.01)), upr_bound=(prop, lambda val:np.quantile(val, q=0.99))).dropna()
    #x, y and sigma data for input to curve fit
    x = df_stats.index
    y = df_stats['mean']
    sig = df_stats['std']
    
    #fit data
    popt_dic2[prop], _pcov = curve_fit(quadratic2, x, y, method='lm')
    
    #find lower bounds
    lwr_y = df_stats['lwr_bound']
    popt_lwr_dic2[prop], _pcov = curve_fit(quadratic2, x, lwr_y, method='lm')
    #find upper bounds
    upr_y = df_stats['upr_bound']
    popt_upr_dic2[prop], _pcov = curve_fit(quadratic2, x, upr_y, method='lm')

    #x-range for lobf and confidence intervals
    x_min=df_temp['Temp (C)'].min()
    x_max=df_temp['Temp (C)'].max()
    x_range=np.arange(x_min,x_max,30)
    
    #plot
    axs[i].plot(df_temp['Temp (C)'], df_temp[prop], marker=".", linewidth=0, label='all data')
    axs[i].plot(x_range, quadratic2(x_range, *popt_dic2[prop]), linestyle="--", color="r", label='fit of mean values')
    axs[i].fill_between(x_range, quadratic2(x_range, *popt_lwr_dic2[prop]), quadratic2(x_range, *popt_upr_dic2[prop]), alpha=0.25, label='confidence interval')
    axs[i].legend()
    axs[i].set_ylabel(prop)
    axs[i].set_xlabel('HTT (ºC)')


plt.show()
fig.savefig('quadratic_fittedplots.png')
```

## linear

```python
#define linear function
def linear(x, a=1, b=0):
    y = a* x + b
    return y

linear_bounds=((0,0),(np.inf,np.inf))
```

```python
linear_properties=['SA (m2g-1)', 'Ash (%)']

#make dictionaries for parameters and errors
popt_dic3={prop:[] for prop in linear_properties}
#make dictionaries for confidence limits
popt_upr_dic3={prop:[] for prop in linear_properties}
popt_lwr_dic3={prop:[] for prop in linear_properties}
```

```python
fig, axs = plt.subplots(nrows=2,figsize=[8,10])

for i, prop in enumerate(linear_properties):   
    df_temp=df[['Temp (C)',prop]].dropna()
    #make temp means, standard devs and standard errors df
    df_stats = df_temp.groupby("Temp (C)").agg(mean=(prop,'mean'), std=(prop,'std'), count=(prop,'count'), lwr_bound=(prop, lambda val:np.quantile(val, q=0.01)), upr_bound=(prop, lambda val:np.quantile(val, q=0.99))).dropna()
    #x, y and sigma data for input to curve fit
    x = df_stats.index
    y = df_stats['mean']
    sig = df_stats['std']
    
    #fit data
    popt_dic3[prop], _pcov = curve_fit(linear, x, y, method='dogbox',bounds=linear_bounds)
    
    #find lower bounds
    lwr_y = df_stats['lwr_bound']
    popt_lwr_dic3[prop], _pcov = curve_fit(linear, x, lwr_y, method='dogbox',bounds=linear_bounds)
    #find upper bounds
    upr_y = df_stats['upr_bound']
    popt_upr_dic3[prop], _pcov = curve_fit(linear, x, upr_y, method='dogbox',bounds=linear_bounds)

    #x-range for lobf and confidence intervals
    x_min=df_temp['Temp (C)'].min()
    x_max=df_temp['Temp (C)'].max()
    x_range=np.arange(x_min,x_max,30)
    
    #plot
    axs[i].plot(df_temp['Temp (C)'], df_temp[prop], marker=".", linewidth=0, label='all data')
    axs[i].plot(x_range, linear(x_range, *popt_dic3[prop]), linestyle="--", color="r", label='fit of mean values')
    axs[i].fill_between(x_range, linear(x_range, *popt_lwr_dic3[prop]), linear(x_range, *popt_upr_dic3[prop]), alpha=0.25, label='confidence interval')
    axs[i].legend()
    axs[i].set_ylabel(prop)
    axs[i].set_xlabel('HTT (ºC)')


plt.show()
fig.savefig('linear_fittedplots.png')
```

## get predicted values

```python
def get_properties(T):
    
    predictions={}
    
    for prop in logistic_properties:
        predictions[prop]=[logistic(T, *popt_dic[prop])]
        predictions[prop].append(fittedlogistic(T, *popt_lwr_dic[prop]))
        predictions[prop].append(fittedlogistic(T, *popt_upr_dic[prop]))
    
    for prop in quadratic_properties[2:]:
        predictions[prop]=[quadratic2(T,*popt_dic2[prop])]
        predictions[prop].append(quadratic2(T,*popt_lwr_dic2[prop]))
        predictions[prop].append(quadratic2(T,*popt_upr_dic2[prop]))

    for prop in linear_properties:
        predictions[prop]=[linear(T,*popt_dic3[prop])]
        predictions[prop].append(linear(T,*popt_lwr_dic3[prop]))
        predictions[prop].append(linear(T,*popt_upr_dic3[prop]))
    
    predictions_df=pd.DataFrame.from_dict(predictions)
    predictions_df.rename(index={0:'mean',1:'lower limit',2:'upper limit'},inplace=True)
    return predictions_df
```

```python
#change T to desired HTT to get predicted properties
T=600

print(f"Predicted properties of biochars produced at HTT of {T}ºC")
get_properties(T)
```
